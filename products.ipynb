```
#practice2
#一個function基本上要只做一件事
#refactor 重構 = 調整程式

import os #operating system作業系統

#讀取檔案
def read_file(filename):
  products = []
  with open(filename, 'r', encoding = 'utf-8')as f:
    for line in f:
      if '商品,價格' in line:
        continue #continue=跳到下一迴，只能用在迴圈，寫在迴圈中很高的位置
      name, price = line.strip().split(',')
      #.strip():去掉換行符號 #.split(',')=遇到,的時候就切割成兩塊，split會將line變成清單
      #把清單中第一樣東西[0]裝進name，第二樣東西[1]裝進price的意思
      products.append([name, price])
    print(products)
    return products

#建立二維清單
def user_input(products):
  products = []
  while True:
    name = input('請輸入商品名稱: ')
    if name == 'q':
      break
    price = input('請輸入商品價格: ')
    price = int(price)
    products.append([name, price])
  print(products)
  print(products[0][0])
  return products

#印出所有購買紀錄
def print_products(products):
  for p in products:
    print(p)#會依序列印小清單出來
    print(p[0], '的價格是', p[1])

#寫入檔案
#with=自動close, f=file檔案
#csv=excel檔 encoding=編碼 utf-8=解決excel中文字亂碼(excel還要做其他更改)
def write_file(filename, products):
  with open(filename, 'w', encoding = 'utf-8')as f:
    f.write('商品,價格\n')
    for p in products:
      f.write(p[0] + ',' + str(p[1]) + '\n')#.write=寫入檔案，只能加string #+:字串跟字串or整數跟整數做合併


#main()當作程式的切入點
def main():
  filename = 'products.csv'
  #檢查檔案(products.csv)在不在
  if os.path.isfile(filename):#只是單純的一行if，不需要重複利用
    print('Yeah!找到檔案了!')
    products = read_file(filename)
    products = user_input(products)
    print_products(products)
    write_file('products.csv', products)
  else:
    print('找不到檔案')
main()
```
